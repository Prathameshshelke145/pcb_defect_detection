```python
from ultralytics import YOLO

# Create a new model from config (NO pre-trained weights)
model = YOLO("yolov8n.yaml")   # n = nano; you can try yolov8s.yaml, m, l, x etc.

results = model.train(
    data="/content/datasets/data.yaml",
    epochs=100,
    imgsz=512,
    batch=16,
    patience=30,
    pretrained=False,
    project="pcb_yolov8_scratch",
    name="yolov8n_pcb_scratch",
    val=True
)
```

```python
from ultralytics import YOLO

model = YOLO("/content/pcb_yolov8_scratch/yolov8n_pcb_scratch/weights/best.pt")

metrics = model.val(
    data="/content/datasets/data.yaml",
    split="val"
)

# Print individual metrics
print("Precision:", metrics.box.mp)     # mean precision
print("Recall:", metrics.box.mr)        # mean recall
print("F1 Score:", metrics.box.f1)      # F1
print("Number of classes:", metrics.box.nc)
```
